**Correlation Analysis**

```python
correlation = df[[' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']].corr()
print(correlation)
```

1: Perfect positive correlation (variables move together exactly).  For instance, as the labour participation rate increases (indicating more people willing to work), the number of employed individuals also rises. This suggests that more job opportunities result in higher workforce participation.

-1: Perfect negative correlation (variables move in opposite directions exactly). For example, there is a negative correlation between the unemployment rate and the number of employed individuals. As unemployment rises, the number of employed people decreases.

0: No correlation.  In this case, if thereâ€™s no correlation between the labour participation rate and the unemployment rate, changes in one do not affect the other, regardless of fluctuations.

```python
correlation = df[[' Estimated Unemployment Rate (%)', ' Estimated Employed', ' Estimated Labour Participation Rate (%)']].corr()
plt.figure(figsize=(8, 4))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

Unemployment Rate over time

```python
mean_unemployment_rate = df.groupby(' Date')[' Estimated Unemployment Rate (%)'].mean()
plt.figure(figsize=(8, 4))
plt.plot(mean_unemployment_rate.index, mean_unemployment_rate.values, marker='o', linestyle='-')
plt.title('Mean Unemployment Rate Over Time')
plt.xlabel(' Date')
plt.ylabel('Mean Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

The graph shows a stable unemployment rate around 10% from July 2019 to February 2020.

Thereâ€™s a sharp increase from March 2020, peaking at around 24% in April or May 2020, likely due to the COVID-19 pandemic.

By June 2020, the rate drops sharply, indicating some recovery, but it remains higher than pre-pandemic levels.

**Region with the highest unemployment rate**

```python
import plotly.express as px
fig = px.bar(df, x='Region', y=' Estimated Unemployment Rate (%)', color='Region', title='Region with the highest unemployment rate')
fig.show()
```

Tripura Region has the highest unemployment rate compare to Sikkim which has the lowest rate by around 1.2%

```python
# Using bar plot
plt.figure(figsize=(4, 2))
fig = px.bar(df, x= 'Month', y= ' Estimated Employed', color= 'Month' , title = 'Estimated Employed People')
fig.show()
```

The bar chart illustrates the estimated number of employed people across different months.

The color gradient from dark purple to yellow represents months from January (1) to December (12).

The highest employment estimates appear in June (Month 6), while months like February and September show relatively lower employment estimates.

The higher employment in June could reflect the recovery from COVID-19 lockdowns. Many regions began reopening businesses and industries, leading to a surge in re-employment as restrictions eased.



```python
# Using Sunburts plot to determine unemployment rate by Region and Area
fig = px.sunburst(df, path=['Area', 'Region'], values=' Estimated Unemployment Rate (%)',
                  title='Unemployment Rate by Region and Area')
fig.show()
```

The chart illustrates that **urban areas** contribute more significantly to the unemployment rate than **rural areas**.

Specific regions like **Tripura, Haryana, Delhi, Bihar,** and **Jharkhand** show up prominently in both urban and rural sections, indicating these regions are particularly affected by unemployment.

**In Summarry**

COVID-19 caused a significant spike in unemployment, with a peak of around 24% during the worst months (March-May 2020), and although there was partial recovery in June, the impact on the job market was substantial.

in the best-case scenario, unemployment could gradually return to pre-pandemic levels within 1-3 years, depending on the country and economic conditions.
